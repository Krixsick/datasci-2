```python
import pandas as pd
import os

# Define the paths to your downloaded files
# (We use os.path.join so it works on Windows and Mac)
matches_path = os.path.join('valorant', 'vct_2025', 'matches', 'overview.csv')
players_path = os.path.join('valorant', 'vct_2025', 'players_stats', 'players_stats.csv')

# Load them
if os.path.exists(matches_path) and os.path.exists(players_path):
    df_matches = pd.read_csv(matches_path)
    df_players = pd.read_csv(players_path)
    print("‚úÖ Success! Data Loaded.")
    print(f"Matches: {len(df_matches)}")
    print(f"Player Stats: {len(df_players)}")
else:
    print(f"‚ùå Files not found. Check that you are in the 'server' folder.")
    print(f"Tried looking here: {os.path.abspath(matches_path)}")
```

```python
# 1. Identify the column name (it's usually 'Map' or 'map_name')
map_col = 'Map' if 'Map' in df_matches.columns else 'map_name'

# 2. Count the maps
map_counts = df_matches[map_col].value_counts()

print("üèÜ Most Played Maps:")
print(map_counts.head(5))

# 3. Show a chart
map_counts.plot(kind='bar', title='Map Popularity', figsize=(10, 5))
```
